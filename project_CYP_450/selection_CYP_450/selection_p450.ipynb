Import to read the tsv file

```python
import pandas as pd
```

Creating a list with all the species present in the database AGORA

```python
agora = []
with open('agora.csv','r') as f:
    line = f.readline()
    while line:
        t = line.replace('\n','').replace('\ufeff','').split('_')
        agora.append(' '.join(t[:2]))
        line = f.readline()
```

Reading file where all of the organisms have the enzyme cytochrome P450

```python
df = pd.read_csv('cytochromep450.tsv',sep='\t')
```

Making a list with the names of the organisms

```python
p450 = df['Name'].tolist()
```

Check how many names do the database AGORA and the file with all the organisms that have the enzyme have in common

```python
# using set intersection
A = set(agora).intersection(set(p450))
print(len(A))
```

Getting all the organisms from AGORA that have the enzyme 

In this code it was also considered partial naming 

```python
# Naming may be partial naming
B = dict()
for x in agora:
    for y in p450:
        if x in y:
            if x not in B.keys():
                B[x]=[]
            B[x].append(y)
B
```
